```python
from mesa import Agent, Model
from mesa.time import RandomActivation
from mesa.space import MultiGrid

import random
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
class MoneyAgent(Agent):
    """An agent with fixed initial wealth."""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.wealth = 1
    
    def step(self):
        # The agent's step will go here.
        print('Agent ' + str(self.unique_id) + ' had ' + str(self.wealth) + ' tokens at the start of this step.')
        if self.wealth == 0:
            return
        other_agent = random.choice(self.model.schedule.agents) #select random agent from list of agents in the model
        other_agent.wealth += 1
        self.wealth -= 1

class MoneyModel(Model):
    """A model with some number of agents."""
    def __init__(self, N):
        self.num_agents = N
        self.schedule = RandomActivation(self)
        # Create agents
        for i in range(self.num_agents):
            a = MoneyAgent(i, self)
            self.schedule.add(a)            
    
    def step(self):
        '''Advance the model by one step'''
        print('stepping..')
        self.schedule.step()
        for agent in self.schedule.agents:
            print('Agent ' + str(agent.unique_id) + ' has ' + str(agent.wealth) + ' tokens at the end of this step.')
```

```python
%%time
all_wealth = []
for j in range(100):
    # Run the model
    model = MoneyModel(10)
    for i in range(10):
        model.step()

    # Store the results
    for agent in model.schedule.agents:
        all_wealth.append(agent.wealth)

plt.hist(all_wealth, bins=range(max(all_wealth)+1))
```

```python
class GridMoneyAgent(Agent):
    """An grid agent with fixed initial wealth."""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.wealth = 1
        
    def move(self):
        possible_steps = self.model.grid.get_neighborhood(
            self.pos,
            moore=True,
            include_center=False)
        new_position = random.choice(possible_steps)
        self.model.grid.move_agent(self, new_position)
        
    def give_money(self):
        cellmates = self.model.grid.get_cell_list_contents([self.pos])
        if len(cellmates) > 1:
            other = random.choice(cellmates)
            other.wealth += 1
            self.wealth -= 1
            
    def step(self):
        self.move()
        if self.wealth > 0:
            self.give_money()
        

class GridMoneyModel(Model):
    """A grid model with some number of agents."""
    def __init__(self, N, width, height):
        self.num_agents = N
        self.grid = MultiGrid(width, height, True)
        self.schedule = RandomActivation(self)

        # Create agents
        for i in range(self.num_agents):
            a = MoneyAgent(i, self)
            self.schedule.add(a)

            # Add the agent to a random grid cell
            x = random.randrange(self.grid.width)
            y = random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))
```

```python
grid_model = GridMoneyModel(50, 10, 10)
for i in range(200):
    grid_model.step()
```

```python
agent_counts = np.zeros((grid_model.grid.width, grid_model.grid.height))
for cell in grid_model.grid.coord_iter():
    cell_content, x, y = cell
    agent_count = len(cell_content)
    agent_counts[x][y] = agent_count
plt.imshow(agent_counts, interpolation='nearest')
plt.colorbar()
```
